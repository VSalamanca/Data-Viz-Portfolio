```
#IMPORT LIBRARIES WE WILL BE USING
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```
lc_data = pd.read_csv('https://raw.githubusercontent.com/VSalamanca/Data-Viz-Portfolio/master/Lending_Club_Data.csv')
```

```
lc_data
```

```
#CREATE A BOXPLOT FOR LOAN AMOUNTS FOR EACH STATE BASED ON HOMEOWNER TYPE

plt.figure(figsize=(30,10))
bplot = sns.boxplot(y='loan_amnt', x='addr_state', 
                 data=lc_data, 
                 width=1,
                 showfliers=False,
                 hue='home_ownership',
                 palette='hls').set(title = 'Loan Amount by State', xlabel = 'State', ylabel = 'Loan Amount' )
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.).set(title= 'HOMEOWNER TYPE')
plt.show()
```

```
#CREATE A SCATTERPLOT WITH LINE REGRESSION TO SHOW RELATIONSHIP BETWEEN LOAN AMOUNT AND ANNUAL INCOME
plt.figure(figsize=(15,10))
sns.regplot(x="annual_inc", y="loan_amnt", data=lc_data);
plt.xlim(0, 1000000)
plt.ylim(0, 45000)
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount')
plt.title('Relationship between Loan Amount and Annual Income')
plt.show()
```
