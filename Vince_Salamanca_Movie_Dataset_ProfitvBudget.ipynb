```
from vega_datasets import data
```

```
movies_data = data.movies()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
Profit = movies_data.Worldwide_Gross - movies_data.Production_Budget
Budget = movies_data.Production_Budget
Rating = movies_data.IMDB_Rating
plot = plt.scatter(Budget, Profit, c=Rating, cmap='viridis')
cbar = plt.colorbar(plot)

plt.title('Profit vs. Budget')
cbar.set_label('IMDB Rating')
plt.xlabel('Budget')
plt.ylabel('Profit')
plt.gcf().set_size_inches((35, 15))    
plt.show()
```
