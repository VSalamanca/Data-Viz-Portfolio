```
import altair as alt
import pandas as pd
```

```
police_shooting = pd.read_csv('https://raw.githubusercontent.com/VSalamanca/Data-Viz-with-Python/master/Fatal_Police_Shootings.csv')
police_shooting
```

```
new_data = police_shooting[['date', 'name','age', 'gender', 'state', 'race', 'signs_of_mental_illness']]

state_shooting = pd.DataFrame(new_data['state'].value_counts())
state_shooting['Percentage'] = state_shooting['state']/state_shooting['state'].sum()
state_shooting.reset_index(level=0, inplace=True)
state_shooting_1 = state_shooting.rename(columns={'index': 'State', 'state': 'Shootings'})

state_shooting_1
```

```
source = state_shooting_1

alt.Chart(source).mark_bar(
    cornerRadiusTopLeft=3,
    cornerRadiusTopRight=3
).encode(
    x='State',
    y='Shootings',
    color='Percentage'
).properties(
    title="# of Shootings in Each State")
```

```
source = new_data

alt.Chart(source).mark_bar(
    cornerRadiusTopLeft=3,
    cornerRadiusTopRight=3
).encode(
    x='signs_of_mental_illness',
    y='race',
    color='race'
).properties(
    title="# of Shootings in Each State")
```
